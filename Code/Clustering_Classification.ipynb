# **terminal towers + flatiron building, train**

````python
tt_train = '/content/drive/MyDrive/Chloe + Juliana/FinalProject/Data/TerminalTowerNY/Train'
````

````python
list_img = os.listdir(tt_train)
print(list_img)
````

````python
import cv2

all_RGBdata_train = [] #intialization (for this creating a list)
for i in range(len(list_img)):
  print(list_img[i])
  img = list_img[i]
  print(img)
  if img == ".DS_Store":
    continue
  image_open = cv2.imread(os.path.join(tt_train, img))
  resize = cv2.resize(image_open, (128,128))
  arr = np.array(resize)
  print(arr.shape)
  print(arr.ndim)
  arr_mean = np.mean(arr, axis=(0,1))
  arr = arr/255.0
  #arr = arr.reshape(1, -1)
  print(arr.shape)
  all_RGBdata_train.append(arr) #A list of all the arr variables, which are the RGB values in 2D formate of each image
  #print(f'[{img}, R={arr_mean[0]:.1f},  G={arr_mean[1]:.1f}, B={arr_mean[2]:.1f} ]')
print(len(all_RGBdata_train)) #204 TT images
````

````python
print(all_RGBdata_train)
````

````python
import cv2
fi_train = '/content/drive/MyDrive/Chloe + Juliana/FinalProject/Data/FlatironBuildingNY/Train'
list_img = os.listdir(fi_train)
print(list_img)
for i in range(len(list_img)):
  print(list_img[i])
  img = list_img[i]
  print(img)
  if img == ".DS_Store":
    continue
  image_open = cv2.imread(os.path.join(fi_train, img))
  resize = cv2.resize(image_open, (128,128))
  arr = np.array(resize)
  print(arr.shape)
  print(arr.ndim)
  arr_mean = np.mean(arr, axis=(0,1))
  arr = arr/255.0
  #arr = arr.reshape(1, -1)
  print(arr.shape)
  all_RGBdata_train.append(arr)
print(len(all_RGBdata_train)) #136 images(FI)
````

````python
import sklearn
kmeans = sklearn.cluster.KMeans(n_clusters=2, random_state=0, n_init="auto")
````

````python
numtrainimgs = len(all_RGBdata_train)
print(numtrainimgs)
````

````python
all_RGBdata_train = np.array(all_RGBdata_train)
all_RGBdata_train = all_RGBdata_train.reshape(numtrainimgs, -1)
print(all_RGBdata_train.shape)
train_model = kmeans.fit(all_RGBdata_train)
train_predictions = kmeans.predict(all_RGBdata_train)
print(train_predictions)
clusters_shape = train_model.cluster_centers_.shape
print(clusters_shape)
````

````python
#https://www.analyticsvidhya.com/blog/2021/06/k-means-clustering-and-transfer-learning-for-image-classification/
x_data = [i for i in range(49152)]
plt.scatter(x_data,train_model.cluster_centers_[0], color = 'red',alpha=0.2,s=70)
plt.scatter(x_data,train_model.cluster_centers_[1] , color = 'blue',alpha=0.2,s=50)
````
